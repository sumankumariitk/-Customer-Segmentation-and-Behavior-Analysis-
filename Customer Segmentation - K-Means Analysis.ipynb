```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv(r'Mall_Customers.csv')
df.head()
```

```python
plt.figure(1, figsize=(16,10))
sns.pairplot(data=df, hue='Gender')
plt.show()
```

```python
plt.figure(1, figsize=(4,4))
sns.countplot(x='Gender', data=df)
plt.show()
```

```python
plt.figure(1, figsize=(16,4))
n = 0 
for x in ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']:
    n += 1
    plt.subplot(1, 3, n)
    plt.subplots_adjust(hspace=0.5 , wspace=0.5)
    sns.distplot(df[x] , bins=10)
    plt.title('Distplot of {}'.format(x))
plt.show()
```

```python
X = df.iloc[:, -2:]
```

```python
km_inertias, km_scores = [], []

for k in range(2, 11):
    km = KMeans(n_clusters=k).fit(X)
    km_inertias.append(km.inertia_)
    km_scores.append(silhouette_score(X, km.labels_))
    

# Elbow Plot for Inertias
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(2, 11), y=km_inertias)
plt.title('Elbow Graph / Inertia Depending on K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()
```

```python
# Silhouette Score Plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(2, 11), y=km_scores)
plt.title('Silhouette Scores Depending on K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
X = df.iloc[:, -3:]
```

```python
km_inertias, km_scores = [], []

for k in range(2, 11):
    km = KMeans(n_clusters=k).fit(X)
    km_inertias.append(km.inertia_)
    km_scores.append(silhouette_score(X, km.labels_))
    
# Elbow Plot for Inertias
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(2, 11), y=km_inertias)
plt.title('Elbow Graph / Inertia Depending on K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show()
```

```python

# Silhouette Score Plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=range(2, 11), y=km_scores)
plt.title('Silhouette Scores Depending on K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
km = KMeans(n_clusters=5).fit(X)
```

```python

# Add cluster labels to the DataFrame
X['label'] = km.labels_

# K-Means visualization on pair of 2 features
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X.iloc[:, 1], y=X.iloc[:, 2], hue=X['label'])
plt.title('K-Means Clustering Visualization')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# K-Means visualization on another pair of 2 features
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X.iloc[:, 0], y=X.iloc[:, 1], hue=X['label'])
plt.title('K-Means Clustering Visualization')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.show()
```

```python
# K-Means visualization on another pair of 2 features
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X.iloc[:, 0], y=X.iloc[:, 2], hue=X['label'])
plt.title('K-Means Clustering Visualization')
plt.xlabel('Age')
plt.ylabel('Annual Income (k$)')
plt.show()
```
